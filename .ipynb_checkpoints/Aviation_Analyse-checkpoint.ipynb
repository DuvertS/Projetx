## Apercu

```python
# Lire le fichier CSV dans un DataFrame
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Chargement du fichier
    
Utilisation de l'encodage "ISO-8859-1" pour éviter les erreurs de décodage.

```python
df = pd.read_csv('AviationData.csv', encoding='ISO-8859-1', low_memory=False)
```

## Dataset Aviation
Le data dossier contient un du National Transportation Safety Board qui comprend des données sur les accidents d'aviation de 1962 à 2023 concernant les accidents de l'aviation civile et des incidents sélectionnés aux États-Unis et dans les eaux internationales.

```python
df.head()
```

## Nettoyage des données


```python
df.columns = [col.strip().lower().replace(".","_") for col in df.columns]
df
```

```python
# Nettoyage des données
## Suppression des colonnes inutiles
cols_to_drop = ['investigation_type', 'latitude', 'longitude', 'airport_code', 'airport_name',
                'far_description', 'event_id', 'publication_date']
df1 = df.drop(columns=cols_to_drop)
df1
```

## Analyse Exploratoire des données (EDA)
Nombre d'accident par année

Nous avons ensuite regardé combien d'accidents ont eu lieu chaque année.

```python
# Conversion des dates en format datetime si ce n'est pas déjà fait
df1['event_date'] = pd.to_datetime(df1['event_date'], errors='coerce')

# Extraction de l'année
df1['year'] = df1['event_date'].dt.year

# Comptage des accidents par an
accidents_by_year = df1.groupby('year').size()

# Affichage du graphique
plt.figure(figsize=(12, 5))
accidents_by_year.plot()
plt.title("Nombre d'accidents aériens par année (1962-2023)")
plt.xlabel("Année")
plt.ylabel("Nombre d'accidents")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Type de dommages les plus fréquents

Nous avons analysé la colonne aircraft_damage, qui indique le type de dommage subi par l’avion 

```python
# Affichage du type de dommage les plus fréquents subis par les avions
plt.figure(figsize=(10, 4))
df1['aircraft_damage'].value_counts().plot(kind='bar')
plt.title("Type de dommages subis par les avions")
plt.xlabel("Type de dommage")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Insight & Interprétation

Nous avons analysé le nombre d'accidents aériens survenus chaque année entre 1962 et 2023.
Le graphique montre une tendance générale à la baisse du nombre d'accidents au fil du temps, notamment après les années 1990. 
Cette diminution peut s'expliquer par plusieurs facteurs :

- Amélioration des normes de sécurité aérienne

- Progrès technologiques dans la conception des avions

- Meilleure formation des pilotes

- Surveillance plus stricte par les autorités aéronautiques

Toutefois, certaines années présentent des pics, qui peuvent être liés à des événements spécifiques ou à des erreurs humaines ou mécaniques particulières.

### L’analyse de la colonne aircraft_damage révèle que les types de dommages les plus fréquents sont :

- Substantial (Substantiel) : l’avion a subi des dégâts importants mais non destructifs.

- Destroyed (Détruit) : l’avion a été irréparable.

Cela suggère que la majorité des incidents entraînent des dégâts majeurs, ce qui souligne la gravité potentielle des accidents aériens, même s’ils ne sont pas toujours mortels.

Cette information est cruciale pour une entreprise souhaitant investir dans l'aviation, car elle peut orienter ses choix vers des modèles d’avions plus robustes et fiables, ou encore renforcer ses protocoles de maintenance.
